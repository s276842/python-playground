```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig, ax = plt.subplots(figsize=(16, 16))
X_LIM = 50
Y_LIM = 50
POINTS = 1000

for i in range(1, 15):
    parab = lambda y: y ** 2 / (2 * i) ** 2 - i ** 2
    inv_par = lambda y: -y ** 2 / (2 * i) ** 2 + i ** 2

    points_par = [(parab(y), y) for y in range(0, POINTS, 2)]
    points_inv_par = [(inv_par(y), y) for y in range(0, POINTS, 2)]

    ax.plot(*zip(*points_par))
    ax.plot(*zip(*points_inv_par))

ax.set_xlim(0, X_LIM)
ax.set_ylim(0, Y_LIM)

ax.set_xticks(range(0, X_LIM, 5))
ax.set_yticks(range(0, Y_LIM, 5))

circle1 = plt.Circle((0, 0), 25, color='r', fill=False)
ax.add_patch(circle1)
plt.show()
```
